```python
x = sp.Symbol('x',real=True)
f = -12*x**4*sp.sin(sp.cos(x)) - 18*x**3+5*x**2 + 10*x - 30
def func(x):
    return -12*x**4*np.sin(np.cos(x)) - 18*x**3+5*x**2 + 10*x - 30
```

### Ищем корни методом бисекции

```python
def bisection(f, lhs, rhs, eps=1e-10):
    mid = (lhs+rhs)/2
    while abs(lhs-rhs) > eps:
        mid = (lhs+rhs)/2
        if f(lhs)*f(mid) > 0: # f(lhs) > 0 and f(mid) > 0 or f(lhs) < 0 and f(mid) < 0
            lhs = mid
        else:
            rhs = mid
    return mid
```

```python
bisection(func, -10, -6)
```

```python
bisection(func, -6, -4)
```

```python
bisection(func, -4, -1)
```

```python
bisection(func, -1, 3)
```

```python
bisection(func, 3, 5)
```

```python
bisection(func, 5, 10)
```
